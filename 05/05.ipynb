```python
with open("input.txt") as f:
    data = f.read().splitlines()
```

```python
ranges = []
ingredients = []

is_range = True

for l in data:
    if not l:
        is_range = False
        continue
        
    if is_range:
        a, b = l.split("-")
        ranges.append((int(a), int(b)))
    else:
        ingredients.append(int(l))

ranges, ingredients
fresh = 0
for i in ingredients:
    for a, b in ranges:
        if a <= i <= b:
            fresh += 1
            break

fresh
```

```python
class AocRange():
    def __init__(self, a, b):
        self.a = a
        self.b = b

    @property
    def size(self):
        return self.b - self.a + 1

    def update(self, new):
        if new.b > self.b:
            self.b = new.b

    def __gt__(self, other):
        return self.a > other.a

    def __repr__(self):
        return f"{self.a} -> {self.b}"
        

ranges = []

for l in data:
    if not l:
        break
        
    a, b = l.split("-")
    ranges.append(AocRange(int(a), int(b)))

sorted_ranges = sorted(ranges)
out = []

while sorted_ranges:
    current_range = sorted_ranges.pop(0)
    print(f"{current_range=}")
    while sorted_ranges and (sorted_ranges[0].a <= current_range.b):
        print("MERGE")
        merge_range = sorted_ranges.pop(0)
        print(merge_range)
        current_range.update(merge_range)

    out.append(current_range)
    print(out)

sum(r.size for r in out)
```
