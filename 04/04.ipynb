```python
from collections import Counter

with open("input.txt") as f:
    data = f.read().splitlines()

data = [list(line) for line in data]
    
x_max = len(data[0])
y_max = len(data)

def get_pos(x, y):
    if not(0 <= x < x_max):
        return None
    if not(0 <= y < y_max):
        return None
    return data[y][x]

def get_neighbours(x, y):
    tot = []

    if get_pos(x, y) != "@":
        return False
    
    for it_y in range(y - 1, y + 2):
        for it_x in range(x - 1, x + 2):
            if it_x == x and it_y == y:
                continue
            d = get_pos(it_x, it_y)
            if d is not None:
                tot.append(d)
    
    n_count = Counter(tot)
    if n_count["@"] < 4:
        return x, y

def get_all_neighbours():
    to_remove = []
    for y in range(0, y_max):
        for x in range(0, x_max):
            neigh = get_neighbours(x, y)
            if neigh:
                to_remove.append(neigh)

    return to_remove

res = 0

while True:
    to_remove = get_all_neighbours()
    if not to_remove:
        break

    res += len(to_remove)

    for r_x, r_y in to_remove:
        data[r_y][r_x] = "."

res
```
