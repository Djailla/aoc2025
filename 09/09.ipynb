```python
with open("example.txt") as f:
    data = f.read().splitlines()
```

```python
from itertools import combinations

comb = list(combinations(data, 2))
```

```python
def surface(a, b):
    x_a, y_a = a.split(",")
    x_b, y_b = b.split(",")

    return (abs(int(x_a) - int(x_b)) + 1) * (abs((int(y_a) - int(y_b))) + 1)
```

```python
max(surface(a, b) for a, b in comb)
```
