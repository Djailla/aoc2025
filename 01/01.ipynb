```python
with open("input.txt") as f:
    data = f.read().splitlines()
```

```python
pos = 50
c = 0

for m in data:    
    op = m[0]
    val = int(m[1:])
    if op == "L":
        pos -= val
    else:
        pos += val
    if pos % 100 == 0:
        c += 1 

print(c)
```

```python
pos = 50
c = 0

for m in data:    
    op = m[0]
    val = int(m[1:])
    if op == "L":
        start = pos - val
        end = pos - 1
        pos -= val
    else:
        start = pos + 1
        end = pos + val
        pos += val

    l = list(range(start, end + 1))
    has_multiple_100 = len([x for x in l if (x % 100 == 0)])

    c += has_multiple_100

print(c)
```
