```python
with open("input.txt") as f:
    data = f.read().splitlines()
data.reverse()

opertors = data.pop(0)

import re
ops = re.split(r"\s{1,}", opertors.strip())

num_list = [[] for _ in range(len(ops))]
for line in data:
    numbers = re.split(r"\s{1,}", line.strip())
    for idx, n in enumerate(numbers):
        num_list[idx].append(int(n))

import math

out = 0
for idx, op in enumerate(ops):
    if op == "+":
        out += sum(num_list[idx])
    elif op == "*":
        out += math.prod(num_list[idx])
        
out
```
