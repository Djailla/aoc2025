```python
from itertools import combinations

with open("example.txt") as f:
    data = f.read().splitlines()
```

```python
sum(max([int(f"{x}{y}") for x, y in combinations(l, 2)]) for l in data)
```

```python
def drop_min(l):
    min_val = min(l)
    l.pop(l.index(min_val))



for line in data:
    l = [int(x) for x in line]
    while(len(l) > 12):
        drop_min(l)
    
    print(l)
```

```python
sum(max([int("".join(c)) for c in combinations(l, 12)]) for l in data)
```
