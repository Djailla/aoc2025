```python
with open("input.txt") as f:
    data = f.read().splitlines()

beams_pos = set([data.pop(0).index("S")])
beams_pos

width = len(data[0])
splits = 0

while data:
    line = data.pop(0)
    if not "^" in line:
        continue

    to_remove = []
    to_add = []
    for idx, el in enumerate(line):
        if el == "^" and idx in beams_pos:
            print(f"Split at {idx}")
            to_remove.append(idx) 
            if idx - 1 > 0:
                to_add.append(idx - 1)
            if idx + 1 < width:
                to_add.append(idx + 1)

            splits += 1

    print(line)
    for i in to_add:
        beams_pos.add(i)
    for i in to_remove:
        beams_pos.remove(i)

splits
```

```python
with open("input.txt") as f:
    data = f.read().splitlines()

routes = {(data.pop(0).index("S"),)}
width = len(data[0])

routes

leng = 0
tot_len = len(data)
while data:
    line = data.pop(0)
    leng += 1
    print(f"{leng}/{tot_len}, {line} routes={len(routes)}")
    to_remove = []
    to_add = []

#    print(routes)
    
    for idx, el in enumerate(line):
        if el == "^":
            for route in routes:
                if int(route[-1]) == idx:
#                    print(f"Split at {idx}")
                    to_remove.append(route)
                    if idx - 1 >= 0:
                        to_add.append(route + (idx - 1, ))
                    if idx + 1 < width + 1:
                        to_add.append(route + (idx + 1, ))
        else:
            for route in routes:
                if int(route[-1]) == idx:
 #                   print(f"Straigth at {idx}")
                    to_remove.append(route)
                    to_add.append(route + (idx, ))
                    

    for i in to_add:
        routes.add(i)
    for i in to_remove:
        routes.remove(i)

routes
```
