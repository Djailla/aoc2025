```python
with open("input.txt") as f:
    data = f.read().splitlines()
```

```python
my_dict = {}

for l in data:
    source, dest = l.split(":")
    destinations = dest.strip().split(" ")
    my_dict[source] = destinations

my_dict
```

```python
def browse_next(el):
    global counter
    for new in my_dict[el]:
        if new == "out":
            counter += 1
        else:
            browse_next(new)

counter = 0
```

```python
browse_next("you")
```
