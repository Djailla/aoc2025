```python
with open("input.txt") as f:
    data = f.read()
```

```python
split_data = data.split(",")

valid_id = {int(f"{d}{d}") for d in range(1, 100000)}
```

```python
tot = []

for d in split_data:
    start, end = d.split("-")
    range_d = set(range(int(start), int(end) + 1))
    for v in valid_id.intersection(range_d):
        tot.append(v)

print(sum(tot))
```
